```python
import os
from synthacc.ground_models import LayerModel
from synthacc.source.mechanism import FocalMechanism
from synthacc.io import axitra
from synthacc.recordings import plot_recordings
```

```python
a_folder = os.path.join('workspaces', 'axitra')
a_params = {'nfreq': 512, 'tl': 20, 'aw': 2, 'xl': 1000000, 'ikmax': 100000}

#src = (11130.0196724578, 0, 5000)
#rcv = (0, 0, 0)

#t0 = 10
```

```python
import eqcatalog
import robspy
#from mapping.geotools.coordtrans import lonlat_to_lambert1972

id_earth = 8285
[eq] = eqcatalog.rob.query_local_eq_catalog_by_id([id_earth])
#[x], [y] = lonlat_to_lambert1972([eq.lon], [eq.lat])
#src = (x, y, eq.depth * 1000)
#print(src)

station_code = 'MOLS'
[station] = robspy.rob.get_stations(station_code)
coords = station.get_coordinates()
#[x], [y] = lonlat_to_lambert1972([coords['longitude']], [coords['latitude']])
#rcv = (x, y, coords['elevation'])
#print(rcv)
```

```python
src = (201660.4062672563, 214688.85291386582, 4000.0)
rcv = (200048.00022554485, 211702.37893739156, 25.0)
```

```python
## Define velocity model
thickness = [200., 400., 700., 1800., 700.]
VP = [1900., 2000., 3500., 4300., 5900.]
VS = [500., 600., 1940., 2390., 3300.]
Rho = [2030., 2045., 2200., 2400., 2600.]
QP = [1000.] * len(VP)
QS = [100.] * len(VS)
lgm = LayerModel(thickness, VP, VS, Rho, QP, QS)
```

```python
## Focal mechanism
## Thierry
fm_rob = FocalMechanism(313.2, 48.1, 64.)
## INERIS
fm_ineris = FocalMechanism(280, 55, -170)

## Moment (in N.m)
moment = 1.795E+12
```

```python
#axitra.write_input(a_folder, a_params, lgm, src, rcv, moment, fm, t0)
```

```python
w = axitra.Wrapper(a_folder, nfreqs=a_params['nfreq'], duration=a_params['tl'],
                   ground_model=lgm)

## Source time function (Stepp, source duration)
stf = (2, 0.033)
t0 = 0
gmt = 'vel'

sg_rob = w.get_recording(src, rcv, moment, fm_rob, stf, gmt, t0)
sg_ineris = w.get_recording(src, rcv, moment, fm_ineris, stf, gmt, t0)
```

```python
sg_obs = robspy.rob.read_from_dp_waveform_server(station_code, 0, 20, id_earth=id_earth)
sg_obs = sg_obs.detrend('linear').restitute(gmt)
sg_obs.components
```

```python
_ = sg_obs.bandpass(0.05,25).plot()
```

```python
plot_recordings([sg_rob, sg_ineris], colors=['r', 'b', 'k'], widths=[3,2,1], size=(10, 5),
                validate=False)
```
