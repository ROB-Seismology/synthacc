```python
import matplotlib.pyplot as plt
import numpy as np

from synthacc.stats import GaussianACF, ExponentialACF, VonKarmanACF
```

```python
gs = GaussianACF()
ex = ExponentialACF()
vk0 = VonKarmanACF(h=0)
vk1 = VonKarmanACF(h=1)

r = np.arange(0, 5, 0.1)

a = 0.8

plt.figure()
plt.plot(r, gs(r, a), label='gs')
plt.plot(r, ex(r, a), label='ex')
plt.plot(r, vk0(r, a), label='vk (H=0)')
plt.plot(r, vk1(r, a), label='vk (H=1)')
plt.legend()
plt.grid()
plt.xlabel('Distance')
plt.ylabel('Correlation')
plt.title('Autocorrelation functions (a=1)')
plt.show()
```
