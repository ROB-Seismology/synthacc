```python
import os

import numpy as np
import pandas as pd
from tqdm import tqdm_notebook

from data import rauw_fault_surface_trace
from rupture import kinematic_rupture_calculator_logic_tree

from synthacc.source.rupture.slip import SlipDistribution
from synthacc.source.rupture.database import ComposedFiniteRuptureDatabase, FlatFileExporter
```

```python
krclt = kinematic_rupture_calculator_logic_tree()
print(len(krclt))
```

```python
## scenario
trace = rauw_fault_surface_trace(proj=3812) ## ETRS89 / Belgian Lambert 2008=)
print(trace.length)
fault_data = {'trace': trace, 'dip': 60}

magnitude = 6.5
rake = -80

krglt = krclt.get_generator(fault_data, magnitude, rake)
```

```python
## Remove db file if you want to start with new database

db = ComposedFiniteRuptureDatabase(os.path.join('workspaces', 'frd', 'cfrd.hdf5'))
print(len(db))
```

```python
n = 5

for i in tqdm_notebook(range(n)):
    g, l = krglt()
    r = g()
    db.add(r, l)

print(len(db))
```

```python
update = True

ffe = FlatFileExporter(db, os.path.join('workspaces', 'frd', 'cfrd.ff'))
print(len(ffe))

for i, k in enumerate(tqdm_notebook(db.keys)):
    ffe.add(k, update=update)

ffe.export()
```
