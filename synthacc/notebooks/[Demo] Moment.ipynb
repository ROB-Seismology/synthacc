# Moment

```python
import os

import numpy as np

from synthacc.io.scardec import read_moment_rate_function
from synthacc.source.moment import TriangularRateGenerator
```

```python
mrf = read_moment_rate_function(os.path.join('data', 'moment', 'fctmoysource_20110312_014715_OFF_EAST_COAST_OF_HONSHU__JAPAN'))
```

```python
print(mrf.moment)
print(mrf.magnitude)
mrf.plot()
```

```python
mf = mrf.get_moment_function()
print(mf.moment)
print(mf.magnitude)
print(mf.get_duration(percentile=5))
mf.plot()
```

```python
nmrf_model = TriangularRateGenerator().get_nmrf(0.1, 6.5, offset=1)
nmrf_model.plot()
```

```python
mrf_model = nmrf_model * mrf.moment
mrf.plot(model=mrf_model)
```
