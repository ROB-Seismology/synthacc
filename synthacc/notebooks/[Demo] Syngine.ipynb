```python
from synthacc.io import syngine
from synthacc.source.moment import mw_to_m0
from synthacc.source.mechanism import FocalMechanism
```

```python
src_depth = 1000
fm = FocalMechanism(90, 60, -45)
mt = fm.get_moment_tensor(mw_to_m0(5.4))
```

```python
r = syngine.get_recording(0, 0, src_depth, mt, 1, 0, duration=60, gmt='dis', components='ZRT')
r.plot()
```

```python
r = syngine.get_recording(0, 0, src_depth, mt, 1, 0, duration=60, gmt='dis', components='ZNE')
r.plot()
```
