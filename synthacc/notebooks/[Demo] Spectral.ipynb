```python
import os

from synthacc.io.esgmd2 import read_cor, read_fas
from synthacc.spectral import plot_fass
from synthacc.recordings import plot_seismograms
```

```python
root_folder = %pwd

fs = os.path.join(root_folder, 'data', 'spectral', '004676xa.cor')
tgt_acc = read_cor(fs)[-1]

fs = os.path.join(root_folder, 'data', 'spectral', '004676xa.fas')
tgt_fas = read_fas(fs)
```

```python
cal_dft = tgt_acc.get_dft()
cal_fas = cal_dft.fas
```

```python
cal_fas.plot(min_amplitude=10**-8, max_amplitude=10)
```

```python
tgt_fas.plot(min_amplitude=10**-8, max_amplitude=10)
```

```python
plot_fass([tgt_fas, cal_fas])
```

```python
plot_fass([tgt_fas, cal_fas], space='loglin')
```

```python
plot_fass([tgt_fas, cal_fas], space='linlin')
```

```python
cal_acc = cal_dft.get_seismogram(tgt_acc.time_delta)
```

```python
plot_seismograms([[tgt_acc, cal_acc]], widths=[[5, 1]], size=(15, 10))
```
