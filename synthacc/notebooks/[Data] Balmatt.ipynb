```python
import folium

from robspy.synthacc import API
```

```python
api = API()
```

```python
s_date='2016-09-01'
e_date='2016-10-31'
region=(4.9, 5.3, 51.0, 51.5)

c = api.get_catalog(s_date, e_date, ke=False)
c = c.query(region)
len(c)
```

```python
for e in c:
    print(e.key, e.time, e.depth, e.magnitude.ml)
```

```python
mols = api.get_station_by_code('MOLS') ## There is only one MOLS (starts in 2008-06-03)
molt = api.get_station_by_code('MOLT') ## There is only one MOLT (starts in 2008-06-03)
dslb = api.get_station_by_code('DSLB', date='2016-09-15')
print(dslb.key)
```

```python
m = folium.Map(location=[51.25, 5.1], zoom_start=12)

for e in c:
    folium.Marker([e.lat, e.lon], popup='<p>%s</p><p>%s</p><p>M=%s</p>' % (e.time, e.depth, e.magnitude.ml)).add_to(m)

folium.Marker([molt.lat, molt.lon], popup='<p>MOLT</p>', icon=folium.Icon(color='red')).add_to(m)
folium.Marker([dslb.lat, dslb.lon], popup='<p>DSLB</p>', icon=folium.Icon(color='red')).add_to(m)

m
```

```python
mols_c = api.get_measured_events(mols, s_date, e_date).query(region)
molt_c = api.get_measured_events(molt, s_date, e_date).query(region)
dslb_c = api.get_measured_events(dslb, s_date, e_date).query(region)

print(mols.key, len(mols_c))
print(molt.key, len(molt_c))
print(dslb.key, len(dslb_c))
```
