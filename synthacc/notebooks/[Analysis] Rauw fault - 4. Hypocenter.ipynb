```python
import matplotlib.pyplot as plt
import numpy as np

from faults import rauw_simple_fault

from synthacc.source.rupture.slip import MaiBeroza2002RFSDC, RandomFieldSDG
from synthacc.source.rupture.hypo import RandomHCC, MaiEtAl2005HCC
```

```python
segment = rauw_simple_fault()
magnitude = 6.5
```

```python
c = MaiBeroza2002RFSDC(100, 100)
g = RandomFieldSDG(c, segment.width, segment.length, magnitude)
sd = g()
sd.plot()
```

```python
hcc1 = RandomHCC()
hcc2 = MaiEtAl2005HCC()
```

```python
hypos1 = [hcc1(segment, sd) for i in range(100)]
hypos2 = [hcc2(segment, sd) for i in range(100)]
```

```python
a = np.where(sd.values >= ((1/3)* sd.max), sd.values, 0)
fig, ax = plt.subplots(figsize=(15, 10))
extent = [0, sd.l/1000, sd.w/1000, 0]
ax.imshow(a, cmap='Greys', extent=extent)
ax.scatter([h.y/1000 for h in hypos1], [h.x/1000 for h in hypos1])
plt.show()

a = np.where(sd.values >= ((1/3)* sd.max), sd.values, 0)
fig, ax = plt.subplots(figsize=(15, 10))
extent = [0, sd.l/1000, sd.w/1000, 0]
ax.imshow(a, cmap='Greys', extent=extent)
ax.scatter([h.y/1000 for h in hypos2], [h.x/1000 for h in hypos2])
plt.show()
```

Hypocenter:
- between 5 and 20 km
- hypocenters are not randomly located on a fault but are located either within or close to regions of large slip (Mai et al., 2005).
