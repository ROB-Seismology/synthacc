```python
import matplotlib.pyplot as plt
import numpy as np

from synthacc.source.rupture.slip import LiuArchuleta2004NSRC, LiuEtAl2006NSRC, SchmedesEtAl2010NSRC
```

```python
d = 0.05
rise_time = 2
times = d * np.arange(np.round(rise_time / d) +1)
```

```python
nsrc1 = LiuArchuleta2004NSRC()
nsrc2 = LiuEtAl2006NSRC()
nsrc3 = SchmedesEtAl2010NSRC()
```

```python
nsrf1 = nsrc1(d, rise_time)
nsrf2 = nsrc2(d, rise_time)
nsrf3 = nsrc3(d, rise_time)

plt.plot(times, nsrf1, label='LiuArchuleta2004')
plt.plot(times, nsrf2, label='LiuEtAl2006')
plt.plot(times, nsrf3, label='SchmedesEtAl2010')
plt.axvline(rise_time/5, c='r')
plt.legend()
plt.grid()
plt.show()
```
