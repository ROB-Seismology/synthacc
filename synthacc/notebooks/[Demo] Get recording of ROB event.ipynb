```python
from robspy.synthacc import API
```

```python
api = API()
```

```python
e = api.get_event(6475)
```

```python
r = api.get_recording('HOU', e.time, e.time+10)
r.plot(title='2018-04-08 Waimes (BE) M1.9 - HOUVEGNEZ (8 km)', size=(15, 10), filespec=None)
```

```python
r = api.get_recording('TNL', e.time+3, e.time+15).filter('highpass', 0.5)
r.plot(title='2018-04-08 Waimes (BE) M1.9 - TERNELL (21 km)', size=(15, 10), filespec=None)
```
