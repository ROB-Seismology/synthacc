```python
import numpy as np

from synthacc.gmpes import GMPE, plot_gmpes_distance, plot_gmpes_magnitude, plot_gmpes_spectrum
```

```python
gmpes = [GMPE('BindiEtAl2017Rjb'), GMPE('AkkarEtAlRjb2014')]

for gmpe in gmpes:
    print(gmpe.tectonic_region)
```

```python
distances = np.logspace(np.log10(1000), np.log10(100000), 25)
```

```python
plot_gmpes_distance(gmpes, 'pga', distances, parameters={'mag': 7, 'rake': 0, 'vs30': 800}, unit='g')
```

```python
plot_gmpes_distance(gmpes, 0.1, distances, parameters={'mag': 7, 'rake': 0, 'vs30': 800}, unit='g')
```

```python
plot_gmpes_distance(gmpes, 1, distances, parameters={'mag': 7, 'rake': 0, 'vs30': 800}, unit='g')
```
