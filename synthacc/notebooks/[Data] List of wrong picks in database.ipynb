```python
from robspy.base import API


api = API()
```

```python
## These are all the picks in our database ('mesure_t' table) from the internal catalog and with a time precision of 8 

picks = api.get_picks()
len(picks)
```

```python
## The wrong ones have a time that is not between that of the station

wrong_picks = []
for p in picks:
    if not (p.eq.begin.date() <= p.date and (type(p.eq.end) is str or p.eq.end is None or p.date <= p.eq.end.date())):
        wrong_picks.append(p)
len(wrong_picks)
```

```python
## These are the stations affected (code = station.place.code + station.code_sup)

stations = set()
for p in wrong_picks:
    stations.add(p.eq.code)

print(stations)
```

```python
## Thise is a list for the picks of DSLB

for p in wrong_picks:
    if p.eq.code == 'DSLB':
        print(p.id_mesure_t, p.date, '(', p.eq.code, p.eq.id_station, p.eq.begin.date(), p.eq.end, ')')
```
